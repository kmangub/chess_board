```python
class Board():
    def __init__(self):
        self.grid = np.zeros((8, 8, 3))
        self.red_coordinates = tuple()
        self.blue_coordinates = tuple()
        self.red = (1, .2, 0)
        self.blue = (0, 1, 1)
        self.white = (1, 1, 1)
        
    def chess_board(self):
        for i in range(4): 
            for j in range(4):  
                self.grid[2*i][2*j] = self.white

        for i in range(4): 
            for j in range(4):   
                self.grid[(2*i)+1][(2*j)+1] = self.white
    
        if self.red_coordinates:
            self.grid[self.red_coordinates] = self.red
        
        if self.blue_coordinates:
            self.grid[self.red_coordinates] = self.blue


    def add_red(self, row, col):
        
        self.grid[row, col] = self.red
        self.red_coordinates = (row,col)
        self.chess_board()

    def add_blue(self, row, col):
        
        self.grid[row, col] = self.blue
        self.blue_coordinates = (row, col)

    def render(self):
        plt.imshow(self.grid)

    def is_under_attack(self):
        if self.red_coordinates[0] == self.blue_coordinates[0] or self.red_coordinates[1] == self.blue_coordinates[1]:
            print(1)
            return True
        elif abs(self.red_coordinates[0] - self.blue_coordinates[0]) == abs(self.red_coordinates[1] - self.blue_coordinates[1]):
            print(2)
            return True
        else:
            print(3)
            return False


    
```

```python
chess_board_test = Board()
chess_board_test.add_red(2,4)
print(chess_board_test.red_coordinates)
chess_board_test.add_blue(4,5)
print(chess_board_test.blue_coordinates)
chess_board_test.render()
chess_board_test.is_under_attack()
```
